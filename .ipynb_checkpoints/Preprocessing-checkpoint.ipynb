```python
import pandas as pd
import matplotlib.pyplot
import nltk
import string
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df = pd.read_csv('dreaddit-train.csv')
```

```python
label = df['label']
text = df['text']
```

```python
df = pd.concat([label, text], axis = 1)
```

```python
df['text'] = df['text'].apply(lambda x: x.split())
```

```python
table  = str.maketrans('', '', string.punctuation)
def clean(bad_array):
    return [word.translate(table) for word in bad_array]
def lower(mixed_array):
    return [word.lower() for word in mixed_array]
def remove_word(horrendous_array, stop_words):
    return [word for word in horrendous_array if not word in stop_words]
```

```python
df['text']  = df['text'].apply(lambda x: clean(x))
df['text'] = df['text'].apply(lambda x: lower(x))
```

```python
from nltk.corpus import stopwords
stop_words= stopwords.words('english')
stop_words  = clean(stop_words)
stop_words = lower(stop_words)
```

```python
df['text'] = df['text'].apply(lambda x: remove_word(x, stop_words))
```

```python
def lemmatize(full_array, lemmatizer):
    return [lemmatizer.lemmatize(word) for word in full_array]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
df['text'] = df['text'].apply(lambda x : lemmatize(x, lemmatizer))
```
